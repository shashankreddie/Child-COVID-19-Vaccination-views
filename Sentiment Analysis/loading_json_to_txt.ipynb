```python
#loading all the json files from a text file and loading tweet data onto a list
import json
whole_data = []
file = open('filename.txt')
lines = file.readlines()
for eachline in lines:
    eachline = eachline.strip('\n')
    print(eachline)
    f = open(eachline)
    json_file = json.load(f)
    whole_data.append(json_file)
    f.close()
file.close()
```

```python
#writing all the data onto a text file for backup
with open('data-1.txt', 'w', encoding='UTF8') as f:
    for item in whole_data:
        f.write("%s\n" % item)
```

```python
import re
mod_data = []
pattern = '[^A-Za-z0-9,: /]+'
with open('data-mod.csv', 'a+', encoding='UTF8') as f:
    for eachline in whole_data:
        mod_string = re.sub(pattern, ' ',str(eachline))
        mod_data.append(mod_string)
        
        
    f.writelines('\n'.join(mod_data))
```
